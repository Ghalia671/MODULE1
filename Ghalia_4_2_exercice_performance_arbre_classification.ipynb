```
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
df = pd.DataFrame(data=cancer['data'], columns = cancer["feature_names"])
df["target"]=cancer['target']
```

Construire des arbres de décision avec des variables de votre choix.


```
df.head()

# exemples de variables à prendre en compte

X=df[["mean radius","mean texture","mean perimeter","mean smoothness"]]
y=df["target"]
```

```
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

dtc = DecisionTreeClassifier(max_depth=2)
dtc.fit(X,y)
from sklearn.tree import plot_tree
plot_tree(dtc,filled=True,impurity=True)
plt.show()
```

**Question** Utiliser la fonction `crosstab` du package `pandas` par exemple pour construire la matrice de confusion.



```
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
X, y = make_classification(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(
        X, y, random_state=0)
clf = SVC(random_state=0)
clf.fit(X_train, y_train)
SVC(random_state=0)
plot_confusion_matrix(clf, X_test, y_test)
plt.show()



```

**Question** Calculer les différentes métriques: justesse (*accuracy*), précision, rappel, score F1.

```
accuracy = 1 - ((2+1)/(12+1+2+10))
print("accuracy : ",round(accuracy * 100,2)," %")

precision = 10/(10+2)
print("precision : ",round(precision * 100,2)," %")

rappel = 10 / (10+2)
print("rappel : ",round(rappel * 100,2)," %")

score_F1 = 2 * ((precision * rappel)/(precision + rappel))
print("score F1 : ",round(score_F1 * 100,2)," %")
```
