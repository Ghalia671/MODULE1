# Introduction

Dans ce notebook, on va tester différents hyperparamètres de `DecisionTreeClassifier`. De façon générale, il est toujours intéressant et important de tester les hyperparamètres d'un algorithme, même si on ne connait pas la signification de tous les hyperparamètres, on peut lire les commentaires dans la documentation de la fonction, et on peut toujours tester différentes valeurs pour voir les effets.

# Profondeur de l'arbre `max_depth`

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
df = pd.DataFrame(data=cancer['data'], columns = cancer["feature_names"])
df["target"]=cancer['target']
```

```
import matplotlib.pyplot as plt

plt.scatter(df["mean radius"], y=df["target"],c=df["target"])
plt.xlabel("mean radius")
plt.ylabel("y")
```

```
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=5)
```

**Question** tester différentes valeurs de `max_depth`

```
X=df[["mean radius"]]
y=df["target"]

dtc.fit(X,y)
```

```
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(dtc,filled=True,impurity=True)
plt.show()
```

# Autres hyperparamètres

**Question** vous pouvez tester différents hyperparamètres suivants:

- `min_samples_split`
- `min_samples_leaf`
- `min_impurity_decrease`
- `min_impurity_split`
- `ccp_alpha`



```
dtc = DecisionTreeClassifier(max_depth=5,min_impurity_decrease=1.0,min_impurity_split=1,min_samples_leaf=2,min_samples_split=4,ccp_alpha=0.0)
X=df[["mean radius"]]
y=df["target"]

dtc.fit(X,y)

plt.figure(figsize=(20,10))
plot_tree(dtc,filled=True,impurity=True)
plt.show()
```

```

dtc = DecisionTreeClassifier()
```

```
X=df[["mean radius"]]
y=df["target"]

dtc.fit(X,y)

plt.figure(figsize=(20,10))
plot_tree(dtc,filled=True,impurity=True)
plt.show()
```
