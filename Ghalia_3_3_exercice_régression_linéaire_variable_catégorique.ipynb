# Apprentissage du modèle

**Question** créez un modèle linéaire avec une variable catégorique, `driveway` par exemple.

Pour cela, vous devez remplacer les valeurs de `yes` et `no` par 0 et 1.

```
X = housing[['driveway']]
y = housing[['price']]
```

```
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns
from sklearn.linear_model import LinearRegression

X["driveway"]= X.apply(lambda row: 1 if row["driveway"] =='yes' else 0, axis=1)

model = LinearRegression()
model.fit(X,y)

model.predict(X) * model.coef_ + model.intercept_
```

**Question** calculer les prix moyens des maisons avec et sans `driveway`.

```
means = y.groupby(X['driveway'])['price'].mean()

means
```

**Question** extraire les coefficients du modèle

```
s = np.array([[0,y]])
a = np.array([[1,y]])
print("Coef pour driveway est 0 : ",model.coef_@s)
print("Coef pour driveway est 1 : ",model.coef_@a)
model.coef_
```

**Question** Comparer les moyennes trouvées aux coefficients.

```
elt1 = s[0][1]
for item1 in elt1['price'] : 
  if item1 > means[0] :
    print("coef supérieur pour 0")
  elif item1 < means[0] :
    print("coef inférieur pour 0")
  else:
    print("égalité")

elt2 = a[0][1]
for item2 in elt2['price'] : 
  if item2 > means[0] :
    print("coef supérieur pour 1")
  elif item2 < means[0] :
    print("coef inférieur pour 1")
  else:
    print("égalité")
```

```
import matplotlib.colors as col
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(y, a[0][1]['price'], s[0][1]['price'], 'grey')
plt.show()
```

**Question** Quel est le RMSE du modèle ? Est-il mieux que le modèle avec la variable `lotsize`?

```
E = model.predict(X) - y
S = E**2
M = S.mean()
R = np.sqrt(M)
R
# il me semble que le modèle est moins précis qu'avec la variable lotsize
```
