**Question** construire un data frame à partir de `data` et `target`.

A la fin, vous devez obtenir un data frame `df` qui contient les variables explicatives avec les noms de colonnes qui correspondent à `cancer["feature_names"]` et on peut appeler `target` la variable cible.

```
df = pd.DataFrame(cancer["data"],columns=cancer["feature_names"])
df["target"] = cancer["target"]
df
```

```
#@title Solution (après avoir essayé, vous pouvez consulter solution), et le data frame qu'il faut trouver est affiché ci-dessous.

df = pd.DataFrame(data=cancer['data'],
                  columns = cancer["feature_names"])
df["target"]=cancer['target']
df
```
