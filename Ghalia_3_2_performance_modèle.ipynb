# Création du modèle: régression linéaire simple

```
import pandas as pd
housing=pd.read_csv("https://dataks.bitbucket.io/ml/houses.csv")
```

```
X = housing[['lotsize']]
y = housing[['price']]
```

```
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```
model.fit(X, y)
```

```
import matplotlib.pyplot as plt

plt.scatter(X,y) # nuage de points
```

```
plt.scatter(X,y) # nuage de points
plt.plot(X,model.predict(X),color="black")
plt.xlabel('lotsize')
plt.ylabel('price')
plt.show()
```

```
model.predict([[400]])
```

```
model.coef_
```

```
model.intercept_
```

```
model.coef_*400+model.intercept_
```

# Performance du modèle

```
model.predict(X)
```

```
residu=model.predict(X)-y
```

```
rescarres=residu**2
```

```
rescarres.mean()
```

```
import numpy as np

np.sqrt(rescarres.mean())
```

```
from sklearn.metrics import mean_squared_error
import numpy as np

lin_mse = mean_squared_error(y, model.predict(X))
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```
from sklearn.metrics import mean_absolute_error

lin_mae = mean_absolute_error(y, model.predict(X))
lin_mae
```

```
ratio_abs=abs((model.predict(X)-y)/y)
```

```
ratio_abs.mean()
```

```
abs((model.predict(X)-y)/model.predict(X)).mean()
```

# Exercices

**Exercice** Implémenter les fonctions pour les différentes suivantes:

- MSE
- MAE
- MAPE



```
import numpy as np
#MSE
som1 = model.predict(X)-y
res1 = som1**2
moy1 = res1.mean()
mse = np.sqrt(moy1.mean())
print("MSE est : ",mse)
#MAE
abs = np.absolute(som1)
res2 = abs**2
moy2 = res2.mean()
mae = np.sqrt(moy2.mean())
print("MAE est : ",mae)
#MAPE
prct1 = abs/y
res3 = prct1**2
moy3 = res3.mean()
mape = np.sqrt(moy3.mean())
print("MAPE est : ",mape)
```
