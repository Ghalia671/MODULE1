# ACP

```
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
```

```
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_r = pca.fit(X).transform(X)
```

```
plt.figure()
plt.scatter(X_r[:, 0], X_r[:, 1])

plt.title('ACP iris')

plt.show()
```

**Question** afficher uniquement la première composante.

```
from sklearn.decomposition import PCA

pca2 = PCA(n_components=1)

X_r1 = pca2.fit(X).transform(X)

plt.figure()
plt.scatter(X_r1[:, 0], X_r1[:, 0])

plt.title('ACP iris 1 Dim')

plt.show()
```

# Visuaisation avec les vraies catégories

```
y = iris.target
target_names = iris.target_names

plt.figure()
colors = ['red', 'blue', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')


plt.show()
```

```
pca.explained_variance_ratio_
```

**Question** visualiser les couleurs des trois espèces sur le graphique à une dimension.

```
plt.figure()
colors = ['red', 'blue', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r1[y == i, 0], X_r1[y == i, 0], color=color, alpha=.8, lw=lw,
                label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset 1 Dim')


plt.show()
```

# ACP appliqué à `breast_cancer`

**Question** appliquer l'ACP à la base `breast_cancer` après avoir normalisé les variables. Visualiser les deux premières composantes principales.

```
from sklearn import datasets

cancer = datasets.load_breast_cancer()

X_c = cancer.data

pca1 = PCA(n_components=2)

X_r3 = pca1.fit(X_c).transform(X_c)

plt.figure()
plt.scatter(X_r3[:, 0], X_r3[:, 1])

plt.title('ACP cancer')

plt.show()
```

```
y_c = cancer.target
target_names = cancer.target_names

plt.figure()
colors = ['red', 'blue', 'green']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r3[y_c == i, 0], X_r3[y_c == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Cancer dataset')


plt.show()
```
