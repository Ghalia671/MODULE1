```
import pandas as pd
housing=pd.read_csv("https://dataks.bitbucket.io/ml/houses.csv")
housing.head(10)
```

# Apprentissage du modèle

**Question** créez un modèle linéaire avec toutes les variables explicatives de la base

```
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd  
import seaborn as sns
from sklearn.linear_model import LinearRegression

housing["driveway"]= housing.apply(lambda row: 1 if row["driveway"] =='yes' else 0, axis=1)
housing["recroom"]= housing.apply(lambda row: 1 if row["recroom"] =='yes' else 0, axis=1)
housing["fullbase"]= housing.apply(lambda row: 1 if row["fullbase"] =='yes' else 0, axis=1)
housing["gashw"]= housing.apply(lambda row: 1 if row["gashw"] =='yes' else 0, axis=1)
housing["airco"]= housing.apply(lambda row: 1 if row["airco"] =='yes' else 0, axis=1)
housing["prefarea"]= housing.apply(lambda row: 1 if row["prefarea"] =='yes' else 0, axis=1)
housing.head(10)
```

```
X = housing[['lotsize','bedrooms','bathrms','stories','driveway','recroom','fullbase','gashw','airco','garagepl','prefarea']]
y = housing[['price']]

model = LinearRegression()
model.fit(X,y)
model.predict(X)
```

# Performance du modèle

```
plt.plot(y,model.predict(X),color="black")
plt.xlabel('price')
plt.ylabel('input')
plt.show()
```

**Question** Quel est le RMSE du modèle ? De combien a-t-on amélioré le modèle par rapport au modèle linéaire simple avec `lotsize`

```
E = model.predict(X) - y
S = E**2
M = S.mean()
R = np.sqrt(M)
R
# on a gagné à peu près 30% de précision par rapport au modèle linéaire simple avec lotsize
```
