# Données

```
import pandas as pd

data=pd.read_csv("https://dataks.bitbucket.io/ml/casa_racine_light.csv")

data
```

Données complètes

```

data_full=pd.read_csv("https://dataks.bitbucket.io/ml/casa_racine_full.csv")
data_full
```

Pour chaque phase de l'apprentisssage, mettre en oeuvre ce que vous avez appris pour créer un modèle à partir des données (simplifiées ou complètes), pour prédire le prix immobilier.



```
import numpy as np

data['Type'] = 1
data['Chambre de service'] = data['Chambre de service'].apply(lambda x: 1 if x == 'Oui' else 0)
data['Mobiliers'] = data['Mobiliers'].apply(lambda x: 1 if x == 'Meublé' else 0)
data['Pièces'] = np.where(
     data['Pièces'] == '3 + Chambre de service', 
    3.5, 
     np.where(
        data['Pièces'] == '4 + Chambre de service', 4.5,
     np.where(
        data['Pièces'] == '5 + Chambre de service', 5.5,data['Pièces']  
    )
  )
)


data_full['Type'] = np.where(
     data_full['Type'] == 'Appartement', 
    1, 
     np.where(
        data_full['Type'] == 'Duplex', 2,
     np.where(
        data_full['Type'] == 'Studio', 3,
     np.where(
        data_full['Type'] == 'Terrain', 4,
        np.where(
        data_full['Type'] == 'Maison', 5,data_full['Type']
            ) 
      )  
    )
  )
)

data_full['Pièces'] = np.where(
     data_full['Pièces'] == '3 + Chambre de service', 
    3.5, 
     np.where(
        data_full['Pièces'] == '4 + Chambre de service', 4.5,
     np.where(
        data_full['Pièces'] == '5 + Chambre de service', 5.5,
        np.where(
        data_full['Pièces'] == 'studio', 0.5,data_full['Pièces']  
    )
  )
)
)

data_full['Mobiliers'] = data_full['Mobiliers'].apply(lambda x: 1 if x == 'Meublé' else 0)

df1 = data_full.dropna()
#df1.isnull().sum()

from sklearn.linear_model import LinearRegression

model_light = LinearRegression()
model_full = LinearRegression()

X1 = data[['Pièces principales','Salles de bains','Type','Pièces','Chambre de service','Mobiliers','Surface']]
y1 = data[['Prix']]

model_light.fit(X1,y1)

X2 = df1[['Salles de bains','Surface','Pièces','Type','Mobiliers']]
y2 = df1[['Prix']]

model_full.fit(X2,y2)
```

# Construction d'un modèle de régression

```
print("modèle data light est : \n",model_light.predict(X1))
print("modèle data full est : \n",model_full.predict(X2))


```
